# Monte Carlo simulations, estimating pi 

### Stochastic vs Deterministic

후자의 경우, 항상 같은 시뮬레이션 결과를 얻을 것이고 전자의 경우, 랜덤니스를 포함하고 있어 매 시뮬레이션 결과가 다름. 

### Static vs __Dynamic__ 

* static : 시간과 관련되있지 않음
* __dynamic__ : 시간과 관련 있음 <br>
ex. Queing networks

### Discrete vs Continuous 

고속도로의 교통흐름을 분석한다고 해보자. <br>
각 자동차 하나하나를 시뮬레이션 할 수 있을 것.(__Discrete__) <br>
아니면 교통흐름을 하나의 흐름(like water)로 보고 시뮬레이션할 수도 있을 것.(Continuous)

우리가 전 시간에 살펴 본 랜덤 워크는 __Stochastic__, __Dynamic__, __Discrete__ 였음.

## Monte Carlo 

랜덤 워크는 몬테카를로의 한 예제이다. (모나코, 몬테카를로의 유명한 카지노가 유래인가..?)

### Inferential statistics(추론 통계)

__Randomly chosen sample__ tends to exhibit the same properties as the population from which it is drawn

### 동전 던지기 

```python
from pylab import *
import random, math

def flipTrial(numFlips):
    heads, tails = 0, 0
    for i in xrange(0, numFlips):
        coin = random.randint(0,1)
        if coin == 0: heads += 1
        else: tails += 1
    return heads, tails

def simFlips(numFlips, numTrials):
    diffs = []
    for i in xrange(0, numTrials):
        heads, tails = flipTrial(numFlips)
        diffs.append(abs(heads-tails))
    diffs = array(diffs)
    diffMean = sum(diffs)/len(diffs)
    diffPercent = (diffs/float(numFlips))*100
    percentMean = sum(diffPercent)/len(diffPercent)
    hist(diffs)
    axvline(diffMean, label = 'Mean')
    legend()
    titleString = str(numFlips) + ' Flips,' + str(numTrials) + ' Trials'
    title(titleString)
    xlabel('Difference between heads and tails')
    ylabel('Number of Trials')
    figure()
    plot(diffPercent)
    axhline(percentMean, label = 'Mean')
    legend()
    title(titleString)
    xlabel('Trial Number')
    ylabel('Percent Difference between heads and tails')
```

```python
simFlips(5000, 100)
figure()
simFlips(100,100)
show()
```

5000번 던지는 것이 차이의 절대량은 늘어나지만 퍼센트는 줄어든다(폭이 much narrower)

```python
simFlips(3,4000)
show()
```
